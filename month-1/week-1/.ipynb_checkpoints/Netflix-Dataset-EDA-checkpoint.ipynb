```python
# Filter only movies
movies_df = df[df["type"] == "Movie"]
movies_df[["type", "duration"]].head(10)
```

```python
TV_Show_df = df[df["type"] == "TV Show"]
TV_Show_df[["type", "duration"]].head(10)
```

```python
movies_df = movies_df.dropna(subset=["duration"])
```

```python
TV_Show_df = TV_Show_df.dropna(subset=["duration"])
```

```python
# Clean movie duration → extract minutes as integer
movies_df["duration_minutes"] = movies_df["duration"].str.extract(r'(\d+)').astype(int)

movies_df[["duration", "duration_minutes"]].head(10)
```

```python
# Clean TV show duration → extract seasons as integer
TV_Show_df["seasons"] = TV_Show_df["duration"].str.extract(r'(\d+)').astype(int)

TV_Show_df[["duration", "seasons"]].head(10)
```

```python
from collections import Counter
import pandas as pd

def get_top_genres(df, n=10):
    """Get top n genres with proper counting"""
    genres = []
    for row in df["listed_in"]:
        if pd.isna(row):  # Handle NaN values
            continue
        for g in row.split(","):
            genres.append(g.strip())
    return Counter(genres).most_common(n)

# Get top genres
top_movie_genres = get_top_genres(movies_df, 10)
top_tv_genres = get_top_genres(TV_Show_df, 10)

# Print formatted results
print("🎬 TOP 10 MOVIE GENRES")
print("=" * 40)
for i, (genre, count) in enumerate(top_movie_genres, 1):
    print(f"{i:2d}. {genre:<30} {count:>4} movies")

print("\n📺 TOP 10 TV SHOW GENRES")
print("=" * 40)
for i, (genre, count) in enumerate(top_tv_genres, 1):
    print(f"{i:2d}. {genre:<30} {count:>4} shows")
```

```python
import matplotlib.pyplot as plt

# Convert to DataFrames for easier plotting
movie_genres_df = pd.DataFrame(top_movie_genres, columns=['Genre', 'Count'])
tv_genres_df = pd.DataFrame(top_tv_genres, columns=['Genre', 'Count'])

# Create bar charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Movie genres chart
ax1.barh(movie_genres_df['Genre'], movie_genres_df['Count'], color='skyblue', edgecolor='navy')
ax1.set_xlabel('Number of Movies')
ax1.set_title('🎬 Top 10 Movie Genres', fontsize=14, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)

# TV show genres chart
ax2.barh(tv_genres_df['Genre'], tv_genres_df['Count'], color='lightcoral', edgecolor='darkred')
ax2.set_xlabel('Number of TV Shows')
ax2.set_title('📺 Top 10 TV Show Genres', fontsize=14, fontweight='bold')
ax2.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```
